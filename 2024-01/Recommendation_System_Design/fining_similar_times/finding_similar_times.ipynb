```python
import csv

movie_titles = {}
genre_sets = {}

with open("ml-25m/movies.csv", "r") as f:
    csvreader = csv.reader(f)
    print(next(csvreader)) # skip cloumn names
    
    for movie_id, title, genre in csvreader:
        movie_titles[int(movie_id)] = title
        genre_sets[int(movie_id)] = set(genre.split("|"))
```

```python
def jaccard_similarity(a, b):
    if len(a | b) == 0: return 0
    return len(a & b) / len(a | b)
```

```python
def get_topk_jaccard_genres(target_moive_id, k=20):
    target_genres_set = genre_sets[target_moive_id]

    result = []

    for movie_id, movie_title in movie_titles.items():
        genresets = genre_sets[movie_id]
        score = jaccard_similarity(target_genres_set, genresets)

        result.append((score, movie_title))

    result.sort(reverse=True)
    return result[:k]
```

```python
movie_id = 164909 # la la land

get_topk_jaccard_genres(movie_id, 20)
```

```python
ratings = []

with open("ml-25m/ratings.csv", "r") as f:
    print(f.readline()) # skip cloumn names

    for line in f:
        user_id, movie_id, rating, timestamp = line.split(",")
        ratings.append((int(user_id), int(movie_id), float(rating)))
```

```python
# jaccard similarity with rating history
from collections import defaultdict

# 만약 비어있다면, 비어있다는 값을 넣기 위해 라이브러리 사용
user_sets = defaultdict(set)

for user_id, movie_id, rating in ratings:
    user_sets[movie_id].add(user_id)
```

```python
# 1번 아이템에 별점을 남긴 유저들의 아이디를 의미한다.
user_sets[1]
```

```python
from tqdm import tqdm

def get_topk_jaccard_ratings(target_moive_id, k=20):
    target_user_set = user_sets[target_moive_id]

    result = []

    for movie_id, movie_title in tqdm(movie_titles.items()):
        usets = user_sets[movie_id]
        score = jaccard_similarity(target_user_set, usets)

        result.append((score, movie_title))

    result.sort(reverse=True)
    return result[:k]
```

```python
movie_id = 112552 # whiplash

get_topk_jaccard_ratings(movie_id, 20)
```

```python
user_rating_sets = defaultdict(dict)

for user_id, movie_id, rating in ratings:
    user_rating_sets[movie_id][user_id] = rating
```

```python
for movie_id, user_rating_set in user_rating_sets.items():
    avg = sum(user_rating_set.values()) / len(user_rating_set)
    
    for k in user_rating_set:
        user_rating_set[k] -= avg
```

```python
def cosine_similarity(a, b):
    numerator = sum(a[k] * b[k] for k in a.keys() & b.keys())
    denominator = (sum(x*x for x in a.values()) * sum(x*x for x in a.values())) ** 0.5
    if denominator == 0: return 0
    return numerator / denominator
```

```python
def get_topk_pearson_ratings(target_moive_id, k=20):
    target_user_rating_set = user_rating_sets[target_moive_id]

    result = []

    for movie_id, user_rating_set in tqdm(user_rating_sets.items()):
        user_rating_set = user_rating_sets[movie_id]
        score = cosine_similarity(target_user_rating_set, user_rating_set)

        result.append((score, movie_id))

    result.sort(reverse=True)
    return result[:k]
```

```python
movie_id = 164909

get_topk_pearson_ratings(movie_id)
```
