```python
print(df_wine[['sulphates', 'alcohol']].shape)
print(df_wine[['sulphates', 'alcohol']][df_wine['quality']>5].shape)
print(df_wine[['sulphates', 'alcohol']][df_wine['quality']<=5].shape)
```

```python
import matplotlib.pyplot as plt

high = df_wine[['sulphates', 'alcohol']][df_wine['quality']>5]
low = df_wine[['sulphates', 'alcohol']][df_wine['quality']<=5]

plt.scatter(high['sulphates'], high['alcohol'], s=50, c='blue', marker='o', label='great')
plt.scatter(low['sulphates'], low['alcohol'], s=50, c='red', marker='v', label='poor')
plt.legend()
plt.grid(True, linestyle='-', alpha=0.3)
plt.show()
```

```python
corr_matrix = df_wine.corr()
corr_matrix['quality'].sort_values(ascending=False)
```
